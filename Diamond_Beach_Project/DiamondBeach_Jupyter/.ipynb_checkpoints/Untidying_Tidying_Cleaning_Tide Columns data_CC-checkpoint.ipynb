```r
library(tidyverse)
library(ggplot2)
```

```r
### Kaimana_tidy dataset
Kaimana_data <- read.csv("Kaimana_tidy - Sheet1.csv")
str(Kaimana_data)
```

```r
### Combining PreTide and PostTide columns into one
Kaimana_data$PreTide_PostTide <- paste(Kaimana_data$PreTide, Kaimana_data$PostTide, sep="_")
str(Kaimana_data)
```

```r
### Removing old columns
Kaimana_data_new <- Kaimana_data[c("Date", "Collection.Time", "Moon", "Time", "BP3", "OC", "People_total", "Pple_msq", "Tide", "PreTide_PostTide")]
str(Kaimana_data_new)
```

```r
### Tidying up the data again (Cleaning)
Kaimana_data_new[c('PreTide', 'PostTide')] <- str_split_fixed(Kaimana_data_new$PreTide_PostTide, '_', 2)
Kaimana_data_new
```

```r
### Removing extra joint column (Cleaning)
Kaimana_data_updated <- Kaimana_data_new[c("Date", "Collection.Time", "Moon", "Time", "BP3", "OC", "People_total", "Pple_msq", "Tide", "PreTide", "PostTide")]
Kaimana_data_updated
```

```r
### Convert Tide column from inches to meters
Kaimana_data_updated$Tide <- Kaimana_data_updated$Tide * 0.3048
Kaimana_data_updated
```

```r
### Changing PreTide and PostTide columns from character to double strings
Kaimana_data_updated$PreTide = as.double(Kaimana_data_updated$PreTide)
Kaimana_data_updated$PostTide = as.double(Kaimana_data_updated$PostTide)
Kaimana_data_updated
```

```r
### Convert PreTide and PostTide columns from inches to meters
Kaimana_data_updated$PreTide <- Kaimana_data_updated$PreTide * 0.3048
Kaimana_data_updated$PostTide <- Kaimana_data_updated$PostTide * 0.3048
Kaimana_data_updated
```
